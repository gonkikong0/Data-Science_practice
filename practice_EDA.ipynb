# Query() Command 
To find/figure out what might be the reason for duplicate columns. <br>
Query() - Query the columns of a DataFrame with a boolean expression.

```python
#Checking an example duplicate
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
# ~ means inverse of the condition
df.reset_index(drop= True)
df = df.loc[~df.duplicated(subset= ['Coaster_Name', 'Location', 'Opening_Date_Clean'])].copy()
```

<b>Parametric v/s Non-Parametric</b>: They are two statistical tests that differ in their assumption about the data. Parametric tests assume that Data follows a specific Distribution, where-as non-Parametric tests are distribution free tests, don't rely on distribution assumptions and are more flexible.
# What is KDE (Kernel Density Estimation)

It is a non-parametric method for estimating the Probability Density Function (PDF) of a random variable by soothing points.


```python
df['Speed_mph'].plot(kind='kde',  title = "Coaster Speed mph")
ax.set_xlabel('Speed mph')
```

# Step 5 -
## Question that needs to be answered with a Data Set.
Q. What are the locations with fastest roller-coasters? (Location Column) [Minimum 10]

```python
df['Location'].value_counts() #Since we don't need "Other"
```

```python
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean') \
    .plot(kind = 'barh', figsize= (10,6), title = 'Average coastspeed by Location')
```
